# Fill in all the missing values present in all the columns in the dataset

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data['Age']=data['Age'].fillna(data['Age'].median())
```

```python
data['Cabin']=data['Cabin'].fillna('unknown')
```

```python
data['Embarked'].mode()[0]
```

```python
data['Embarked']=data['Embarked'].fillna(data['Embarked'].mode()[0])
```

```python
data.isna().sum()
```

```python
data=data.drop(['Name','Ticket','Cabin'],axis=1)
```

# One Hot Encoding

```python
df=pd.get_dummies(data)
```

```python
df
```

# MinMax Scaling

```python
from sklearn.preprocessing import MinMaxScaler
minmax=MinMaxScaler(feature_range=(0,1)) 
X1=minmax.fit_transform(df)
```

```python
df.columns
```

```python
X1=pd.DataFrame(X1,columns=['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex_female',
       'Sex_male', 'Embarked_C', 'Embarked_Q', 'Embarked_S'])
```

```python
X1
```

```python
X1.index=data.index
```

```python
X1
```

# Correlation Matrix

```python
corrmatrix=X1.corr()
plt.subplots(figsize=(20,8))
plt.title("Correlation matrix \n", size = 20)
sns.heatmap(corrmatrix,vmax=0.7,vmin=-0.7,annot=True,cmap='YlGnBu')
```

# Splitting target and independent features

```python
from sklearn.model_selection import train_test_split
y = X1['Survived']
x = X1.drop('Survived',axis=1)

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#KNN
```

```python
from sklearn.neighbors import KNeighborsClassifier
metric_k= []
neighbors = np.arange(3,15)
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score,accuracy_score

for k in neighbors:
    knn=KNeighborsClassifier(n_neighbors=k)
    model_knn=knn.fit(x_train,y_train)
    y_pred_knn=model_knn.predict(x_test)
    acc=accuracy_score(y_test,y_pred_knn)
    metric_k.append(acc)
```

```python
plt.plot(neighbors, metric_k,'o-')
plt.xlabel('k value')
plt.ylabel('accuracy')
plt.grid()
```
